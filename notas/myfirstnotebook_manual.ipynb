# what we should know about cryptography 

## existen dos metodos de cifrado mas importante los cuales detallaremos a continuacion 
* Cifrado simetrico: Se refiere al conjunto de métodos que permiten tener una comunicación segura entre emisor y receptor, siempre y cuando con anticipación se hayan intercambiado la clave que es llamada clave simétrica. Es denominado cifrado simétrico, porque utiliza la misma clave para cifrar y descifrar un mensaje. Es un metodo sencillo, pero presenta ciertas dificultades: para que el emisor y el receptor puedan leer el mensaje, ambos deben tener la misma clave, y para eso se la debieron haber compartido previamente mediante papel, mensaje, llamado telefonico. No es tan segura como el cifrado asimetrico ya que cualquier persona que tenga la clave (ya sea ses haya filtrado, o es un hacker) puede leer el mensaje. 
* La criptografía simétrica ofrece mucha velocidad en la transmisión y entrega de los mensajes. Sin embargo, como mencionamos antes, el método de protección de la clave es fácilmente vulnerable.
* Cifrado Asimetrico: A diferencia de la criptografía simétrica, la asimétrica tiene dos claves, un pública y una privada. La clave pública no tiene una barrera de protección, sin embargo, la segunda clave es privada y cifra el mensaje, por lo tanto, es el receptor el que cifra el mensaje. Las claves asimétricas tienen un código de encriptación que hace difícil romper el código. La clave privada unicamente la sabe el propietario de la misma.
